```python
import os, sys 
from pathlib import Path
from os.path import dirname, realpath
script_dir = Path(dirname(realpath('.')))
module_dir = str(script_dir)
sys.path.insert(0, module_dir + '/modules')
import numpy as np
import utility as ut
import matplotlib.pyplot as plt
import pandas as pd
import warnings
from scipy import stats
import seaborn as sns
import oneshot as sm
import surrogate as sr 
import torch
import hr_full as hrf
from torch import nn
import eulerN as arch
```

```python
description = "This experiment checks if p_g=0,1 and uniformly sampled RFMs for a function approximation task. The function used here is tasmania(x) = 0.03 - (1.4 - 2.5*(x+0.5)) * np.sin(15*(x+0.5))."
experimentFolder = '../data/pg-0-1-uniform-tasmania'
ut.makeDir(experimentFolder)

tasmania = lambda x: 0.03 - (1.4 - 2.5*(x+0.5)) * np.sin(15*(x+0.5))
a, b = -0.5, 0.5
x = np.linspace(a, b, num=1000000)
x_ = np.random.choice(x, size=50)


test_x = np.random.uniform(a, b, size=(1, 1000))
test_y = tasmania(test_x)

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
ax.scatter(x, tasmania(x), s=1, label="data")

L0, L1, D, D_r, beta, N = 0.4, 3.5, 1, 3000, 1e-10, 2000
Uo = x_.reshape(-1, 1).T
sampler = sm.MatrixSampler(L0, L1, Uo)

saturatedWb = sampler.sample_([0, 0, D_r])
saturatedRFM = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=saturatedWb[:, :-1], b_in_fn=saturatedWb[:, -1], name="Saturated RFM")
saturatedRFM.compute_W_xy(Uo, tasmania(Uo), beta)
saturatedRFM.validation_error(test_x, test_y)

goodWb = sampler.sample_([D_r, 0, 0])
goodRFM = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=goodWb[:, :-1], b_in_fn=goodWb[:, -1], name="One shot Good RFM")
goodRFM.compute_W_xy(Uo, tasmania(Uo), beta)
goodRFM.validation_error(test_x, test_y)

uniformWb = np.random.uniform([-5., -2.], [5., 2.], size=(D_r, D+1))
uniformRFM = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=uniformWb[:, :-1], b_in_fn=uniformWb[:, -1], name="Uniform RFM")
uniformRFM.compute_W_xy(Uo, tasmania(Uo), beta)
uniformRFM.validation_error(test_x, test_y)

goodWb = hrf.GoodRowSampler(L0, L1, Uo.T).sample(D_r, steps=10)
goodRFMFull = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=goodWb[0], b_in_fn=goodWb[1], name="Bisection Good RFM")
goodRFMFull.compute_W_xy(Uo, tasmania(Uo), beta)
goodRFMFull.validation_error(test_x, test_y)

goodWb = hrf.GoodRowSampler(0., L1, Uo.T).sample(D_r, steps=10)
glRFMFull = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=goodWb[0], b_in_fn=goodWb[1], name="Bisection Linear Good RFM")
glRFMFull.compute_W_xy(Uo, tasmania(Uo), beta)
glRFMFull.validation_error(test_x, test_y)

goodWb = hrf.GoodRowSampler(0., 0.4, Uo.T).sample(D_r, steps=10)
glsRFMFull = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=goodWb[0], b_in_fn=goodWb[1], name="Bisection Linear Good Saturated RFM")
glsRFMFull.compute_W_xy(Uo, tasmania(Uo), beta)
glsRFMFull.validation_error(test_x, test_y)

x = np.linspace(a, b, num=1000)
Ux = x.reshape(-1, 1).T

ax.plot(x, saturatedRFM.forecast_m(Ux).flatten(), label=f"{saturatedRFM.name}")
ax.plot(x, goodRFM.forecast_m(Ux).flatten(), label=f"{goodRFM.name}")
ax.plot(x, uniformRFM.forecast_m(Ux).flatten(), label=f"{uniformRFM.name}")
ax.plot(x, goodRFMFull.forecast_m(Ux).flatten(), label=f"{goodRFMFull.name}")
ax.plot(x, glRFMFull.forecast_m(Ux).flatten(), label=f"{glRFMFull.name}")
ax.plot(x, glsRFMFull.forecast_m(Ux).flatten(), label=f"{glsRFMFull.name}")
ax.scatter(Uo, tasmania(Uo), marker="x")
ax.legend()

expLogger = ut.ExperimentLogger(experimentFolder, description)
expLogger.add(saturatedRFM, goodRFM, uniformRFM, goodRFMFull, glRFMFull, glsRFMFull)

plt.savefig(f"{experimentFolder}/fit_beta_{beta}.png")
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
def scatter_wb(rfm, ax):
    ax.scatter(rfm.W_in.flatten(), rfm.b_in.flatten(), s=5)
    ax.set_xlabel(r"$w_1$")
    ax.set_ylabel(r"$b$")
    ax.set_title(f"{rfm.name}")
scatter_wb(saturatedRFM, axs[0, 0])
scatter_wb(goodRFM, axs[0, 1])
scatter_wb(uniformRFM, axs[0, 2])
scatter_wb(goodRFMFull, axs[1, 0])
scatter_wb(glRFMFull, axs[1, 1])
scatter_wb(glsRFMFull, axs[1, 2])
plt.savefig(f"{experimentFolder}/scatter-wb.png")
```

```python
description = "This experiment checks the forecast times for p_g=0, 1 and uniformly sampled RFMs for L63."
experimentFolder = '../data/pg-0-1-uniform-L63'
ut.makeDir(experimentFolder)

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)

L0, L1, D, D_r, beta, N = 0.4, 3.5, 3, 1024, 4e-5, 20000
L63_data_folder = '../data/L63-trajectories'
Uo = np.load('{}/train.npy'.format(L63_data_folder))[:N]
test = np.load('{}/test.npy'.format(L63_data_folder))[:, :800, :]
sampler = sm.MatrixSampler(L0, L1, Uo)


saturatedWb = sampler.sample_([0, 0, D_r])
saturatedRFM = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=saturatedWb[:, :-1], b_in_fn=saturatedWb[:, -1], name="Saturated RFM")
saturatedRFM.compute_W(Uo, beta)
saturatedTau_f = saturatedRFM.compute_forecast_time(test[:500], 0.05, 0.02)[0]

goodWb = sampler.sample_([D_r, 0, 0])
goodRFM = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=goodWb[:, :-1], b_in_fn=goodWb[:, -1], name="One shot Good RFM")
goodRFM.compute_W(Uo, beta)
goodTau_f = goodRFM.compute_forecast_time(test[:500], 0.05, 0.02)[0]

uniformWb = np.random.uniform([-.4, -.4, -.4, -4.], [.4, .4, .4, 4.], size=(D_r, D+1))
uniformRFM = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=uniformWb[:, :-1], b_in_fn=uniformWb[:, -1], name="Uniform RFM")
uniformRFM.compute_W(Uo, beta)
uniformTau_f = uniformRFM.compute_forecast_time(test[:500], 0.05, 0.02)[0]


goodWb = hrf.GoodRowSampler(L0, L1, Uo.T).sample(D_r, steps=10)
goodRFMFull = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=goodWb[0], b_in_fn=goodWb[1], name="Bisection Good RFM")
goodRFMFull.compute_W(Uo, beta)
goodFullTau_f = goodRFMFull.compute_forecast_time(test[:500], 0.05, 0.02)[0]


ax.hist(saturatedTau_f, label=f"{saturatedRFM.name}", alpha=0.7)
ax.hist(goodTau_f, label=f"{goodRFM.name}", alpha=0.7)
ax.hist(uniformTau_f, label=f"{uniformRFM.name}", alpha=0.7)
ax.hist(goodFullTau_f, label=f"{goodRFMFull.name}", alpha=0.7)
ax.legend()

expLogger = ut.ExperimentLogger(experimentFolder, description)
expLogger.add(saturatedRFM, goodRFM, uniformRFM)

plt.savefig(f"{experimentFolder}/tau_f_beta_{beta}.png")
```

```python
description = "This experiment checks deep vs shallow RFM for approximation task. The function used here is tasmania(x) = 0.03 - (1.4 - 2.5*(x+0.5)) * np.sin(15*(x+0.5))."
experimentFolder = '../data/deep-vs-shallow-tasmania'
ut.makeDir(experimentFolder)

tasmania = lambda x: 0.03 - (1.4 - 2.5*(x+0.5)) * np.sin(15*(x+0.5))
a, b = -0.5, 0.5
x = np.linspace(a, b, num=1000000)
x_ = np.random.choice(x, size=50)


test_x = np.random.uniform(a, b, size=(1, 1000))
test_y = tasmania(test_x)

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
ax.scatter(x, tasmania(x), s=1, label="data")

L0, L1, D, D_r, beta, N = 0.4, 3.5, 1, 300, 1e-10, 2000
Uo = x_.reshape(-1, 1).T
sampler = sm.MatrixSampler(L0, L1, Uo)
B = 5

deepRFM = arch.DeepRF(D_r, B, L0, L1, Uo, beta, name=f"Deep RFM of depth {B}", save_folder=experimentFolder)
deepRFM.init_xy(Uo, tasmania(Uo))

goodWb = sampler.sample_([D_r, 0, 0])
goodRFM = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=goodWb[:, :-1], b_in_fn=goodWb[:, -1], name="One shot Good RFM")
goodRFM.compute_W_xy(Uo, tasmania(Uo), beta)
goodRFM.validation_error(test_x, test_y)

uniformWb = np.random.uniform([-5., -2.], [5., 2.], size=(D_r, D+1))
uniformRFM = sr.SurrogateModel_LR(D=D, D_r=D_r, W_in_fn=uniformWb[:, :-1], b_in_fn=uniformWb[:, -1], name="Uniform RFM")
uniformRFM.compute_W_xy(Uo, tasmania(Uo), beta)
uniformRFM.validation_error(test_x, test_y)

x = np.linspace(a, b, num=1000)
Ux = x.reshape(-1, 1).T

ax.plot(x, saturatedRFM.forecast_m(Ux).flatten(), label=f"{saturatedRFM.name}")
ax.plot(x, goodRFM.forecast_m(Ux).flatten(), label=f"{goodRFM.name}")
ax.plot(x, uniformRFM.forecast_m(Ux).flatten(), label=f"{uniformRFM.name}")

ax.scatter(Uo, tasmania(Uo), marker="x")
ax.legend()

expLogger = ut.ExperimentLogger(experimentFolder, description)
expLogger.add(goodRFM, uniformRFM)

plt.savefig(f"{experimentFolder}/fit_beta_{beta}.png")
```
